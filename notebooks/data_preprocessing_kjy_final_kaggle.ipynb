```python
annotations_path = '/kaggle/working/pill_data/pill_images (1)/train_annotations'

categories_list = []
seen_category_ids = set()

print(f"'{annotations_path}' í´ë”ì—ì„œ í´ë˜ìŠ¤ ì •ë³´ë¥¼ ìˆ˜ì§‘í•©ë‹ˆë‹¤...")

if os.path.exists(annotations_path):
    for root, _, files in tqdm(os.walk(annotations_path), desc="Scanning JSONs"):
        for file in files:
            if file.endswith('.json'):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        if 'categories' in data:
                            for cat in data['categories']:
                                if cat['id'] not in seen_category_ids:
                                    categories_list.append(cat)
                                    seen_category_ids.add(cat['id'])
                except Exception:
                    continue
    
    categories_df = pd.DataFrame(categories_list)
    print(f"\nâœ… {len(categories_df)}ê°œì˜ í´ë˜ìŠ¤ ì •ë³´ë¥¼ ì„±ê³µì ìœ¼ë¡œ ìˆ˜ì§‘í–ˆìŠµë‹ˆë‹¤.")
else:
    print(f"â—ï¸ ì–´ë…¸í…Œì´ì…˜ í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {annotations_path}")
```

```python
base_dir = '/kaggle/working/pill_data/pill_images (1)'
train_img_path = os.path.join(base_dir, 'images/train_images')
val_img_path = os.path.join(base_dir, 'images/val_images')

if 'categories_df' in locals() and not categories_df.empty:
    sorted_categories = categories_df.sort_values(by='id').reset_index(drop=True)
    class_names = sorted_categories['name'].tolist()

    data_yaml = {
        'train': os.path.abspath(train_img_path),
        'val': os.path.abspath(val_img_path),
        'nc': len(class_names),
        'names': class_names
    }

    yaml_path = '/kaggle/working/pill_data/dataset.yaml'
    with open(yaml_path, 'w', encoding='utf-8') as f:
        yaml.dump(data_yaml, f, allow_unicode=True, default_flow_style=False)

    print(f"âœ… ìµœì¢… '{yaml_path}' íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print("\n--- ìƒì„±ëœ íŒŒì¼ ë‚´ìš© (ìƒìœ„ 10ê°œ í´ë˜ìŠ¤) ---")
    for i, name in enumerate(class_names[:10]):
        print(f"  - ID {i}: {name}")
    print("\nYAML íŒŒì¼ì„ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ í›ˆë ¨ì„ ì‹œì‘í•  ì¤€ë¹„ê°€ ë˜ì—ˆìŠµë‹ˆë‹¤.")
else:
    print("â—ï¸ 'categories_df' ë³€ìˆ˜ê°€ ì—†ìŒ.")
```

```python
import yaml
import os

# --- 1. YAML íŒŒì¼ ê²½ë¡œ ì§€ì • ë° ë‚´ìš© í™•ì¸ ---
base_dir = '/kaggle/working/pill_data'
yaml_path = os.path.join(base_dir, 'dataset.yaml')

print(f"--- '{yaml_path}' íŒŒì¼ ë‚´ìš© í™•ì¸ ---")

try:
    with open(yaml_path, 'r', encoding='utf-8') as f:
        data_yaml = yaml.safe_load(f)
        
        print("\n[í˜„ì¬ YAML íŒŒì¼ ë‚´ìš©]")
        print(yaml.dump(data_yaml, allow_unicode=True, default_flow_style=False))

        # --- 2. ê²½ë¡œ ìœ íš¨ì„± ê²€ì¦ ë° ìˆ˜ì • ---
        print("\n--- ê²½ë¡œ ìœ íš¨ì„± ê²€ì¦ ---")
        train_path = data_yaml.get('train', '')
        val_path = data_yaml.get('val', '')
        
        # val ê²½ë¡œê°€ train ê²½ë¡œì™€ ë™ì¼í•œì§€ í™•ì¸
        if train_path == val_path:
            print("â—ï¸ ê²½ê³ : 'val' ê²½ë¡œê°€ 'train' ê²½ë¡œ ìˆ˜ì •ì´ í•„ìš”í•©ë‹ˆë‹¤.")
            
            # ì˜¬ë°”ë¥¸ val ê²½ë¡œ ìƒì„±
            correct_val_path = os.path.abspath(os.path.join(base_dir, 'images/val_images'))
            
            # ê²½ë¡œ ìˆ˜ì •
            data_yaml['val'] = correct_val_path
            
            # ìˆ˜ì •ëœ ë‚´ìš©ìœ¼ë¡œ íŒŒì¼ ë‹¤ì‹œ ì €ì¥
            with open(yaml_path, 'w', encoding='utf-8') as f:
                yaml.dump(data_yaml, f, allow_unicode=True, default_flow_style=False)
            
            print(f"\nâœ… 'val' ê²½ë¡œë¥¼ '{correct_val_path}' (ìœ¼)ë¡œ ìˆ˜ì •í•˜ê³  íŒŒì¼ì„ ë‹¤ì‹œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
            print("\n[ìˆ˜ì •ëœ YAML íŒŒì¼ ë‚´ìš©]")
            print(yaml.dump(data_yaml, allow_unicode=True, default_flow_style=False))
        else:
            print("âœ… 'val' ê²½ë¡œê°€ 'train' ê²½ë¡œì™€ ë‹¤ë¥´ê²Œ ì˜¬ë°”ë¥´ê²Œ ì„¤ì •ë˜ì–´ ìˆìŠµë‹ˆë‹¤.")

except FileNotFoundError:
    print(f"â—ï¸ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {yaml_path}")
except Exception as e:
    print(f"â—ï¸ íŒŒì¼ì„ ì²˜ë¦¬í•˜ëŠ” ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
```
